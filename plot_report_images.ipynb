```python
gtarray = mask
overseg = watr
undrseg = otsu
```

```python
def relu(array):
    output = np.zeros(array.shape)
    output[array>0] = array[array>0]
    return output

def inv_relu(array):
    output = np.zeros(array.shape)
    output[array==1] = 1
    return output
```

```python
t = 135
v1 = gtarray[t:t+95]
v2 = label(overseg[t:t+95])
v3 = gtarray[25:t]
v4 = label(undrseg[25:t])

unq,counts = np.unique(v2,return_counts=True)

v1 = np.stack([v1*0,v1*200,v1*0],axis=-1)
v2 = np.stack([(inv_relu(v2))*200,v2*0,(relu(v2-1))*200],axis=-1)
v3 = np.stack([v3*0,((v3**1)*100*2**0).astype(int),v3*0],axis=-1)
v4 = np.stack([v4*0,v4*0,v4*200],axis=-1)
```

```python
import numpy as np

def replace_black_pixels(rgb_array):
    white_pixel = np.array([255, 255, 255], dtype=np.uint8)
    modified_array = np.where(np.all(rgb_array == [0, 0, 0], axis=2)[:, :, np.newaxis], white_pixel, rgb_array)
    return modified_array
```

```python
fig,axs = plt.subplots(1,6,figsize=[12,5])
axs[0].imshow(replace_black_pixels(v1))
axs[1].imshow(replace_black_pixels(v2))
axs[2].imshow(replace_black_pixels(v1+v2))
axs[3].imshow(replace_black_pixels(v3))
axs[4].imshow(replace_black_pixels(v4))
axs[5].imshow(replace_black_pixels(v3+v4))
axs[0].set_title("Ground Truth")
axs[1].set_title("Over-segmentation")
axs[2].set_title("Overlay")
axs[3].set_title("Ground Truth")
axs[4].set_title("Under-segmentation")
axs[5].set_title("Overlay")
fig.tight_layout()
plt.savefig("report_images/iou_examples.pdf")
plt.show()
```
